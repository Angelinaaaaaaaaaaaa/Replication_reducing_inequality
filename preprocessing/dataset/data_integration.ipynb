```python
import pandas as pd 
import numpy as np
```

```python
df_poll = pd.read_csv('outcome-df/df_poll.csv')
df_poll
```

```python
reg = pd.read_csv('ga_n_registered_race_gender.csv')
reg
```

```python
df_poll.columns
```

```python
reg = reg[['country', 'Reg #']]
# Perform the join on 'country'
df_poll = df_poll.merge(reg, left_on='County', right_on='country', how='inner')

# Create the new column
df_poll['poll'] = df_poll['n_poll'] / df_poll['Reg #']

# Drop the duplicate 'country' column from reg (optional)
df_poll.drop(columns=['country'], inplace=True)

# Display the modified DataFrame
df_poll.head()
```

```python
df_poll.describe()
```

```python
# Extract 'County' values where 'poll' > 0.001 and store as an array
county_array = df_poll[df_poll['poll'] > 0.001]['County'].to_numpy()

# Display the array
print(county_array)
```
